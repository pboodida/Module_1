```python
c = chipo.groupby('item_name')
c = c.sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

```python
c = chipo.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending=False)
c.head(1)
```

```python
total_items_orders = chipo.quantity.sum()
total_items_orders
```

```python
chipo.item_price.dtype
```

```python
dollarizer = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(dollarizer)
```

```python
chipo.info()
```

```python
revenue = (chipo['quantity']* chipo['item_price']).sum()
print('Revenue was: $' + str(np.round(revenue,2)))
```

```python
orders = chipo.order_id.value_counts().count()
orders
```

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
```

```python
order_grouped = chipo.groupby(by=['order_id']).sum(numeric_only=True)
average_revenue = order_grouped.mean()['revenue']
print(average_revenue)
```

```python
# Group by the item name and sum the quantities
item_quants = chipo.groupby('item_name')['quantity'].sum()

# Sort from highest to lowest to see the most popular items
item_quants = item_quants.sort_values(ascending=False)

# Display the top 5
print(item_quants.head(5))
```
