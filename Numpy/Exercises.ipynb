````python
#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)
P0 = np.random.random((5,2))
P1 = np.random.random((5,2))
p  = np.random.random(2)
# Direction vectors of lines
V = P1 - P0              # (5,2)
# Vector from line start to point
W = p - P0               # (5,2)
# Manual 2D cross product
cross = V[:,0]*W[:,1] - V[:,1]*W[:,0]
# Distance formula
dist = np.abs(cross) / np.linalg.norm(V, axis=1)
print(dist)
````

````python
#### 80. Consider an arbitrary array, write a function that extracts a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)
def extract_centered(arr, center, shape, fill=0):
    arr = np.asarray(arr)
    center = np.asarray(center)
    shape = np.asarray(shape)
    # Compute half window sizes
    half = shape // 2
    # Start & end indices in original array
    start = center - half
    end   = start + shape
    # Amount of padding needed
    pad_before = np.maximum(0, -start)
    pad_after  = np.maximum(0, end - arr.shape)
    # Clip slicing indices to valid range
    start = np.maximum(0, start)
    end   = np.minimum(arr.shape, end)
    # Slice the valid part
    sliced = arr[tuple(slice(s, e) for s, e in zip(start, end))]
    # Pad if needed
    if np.any(pad_before) or np.any(pad_after):
        pad_width = list(zip(pad_before, pad_after))
        sliced = np.pad(sliced, pad_width, constant_values=fill)
    return sliced
Z = np.arange(25).reshape(5,5)
result = extract_centered(Z, center=(0,0), shape=(3,3), fill=-1)
print(result)
````

````python
#### 88. How to implement the Game of Life using numpy arrays? (★★★)
def game_of_life(Z):
    N = sum(np.roll(np.roll(Z,i,0),j,1)
            for i in (-1,0,1)
            for j in (-1,0,1)
            if (i,j)!=(0,0))
    return (N==3) | (Z & (N==2))
Z = np.random.randint(0,2,(5,5))
print(game_of_life(Z))
````

````python
#### 99. Given an integer n and a 2D array X, select from X the rows which can be interpreted as draws from a multinomial distribution with n degrees, i.e., the rows which only contain integers and which sum to n. (★★★)
def multinomial_rows(X, n):
    """
    Select rows that:
    1) contain only integers
    2) sum to n
    """
    # Check integer condition
    integer_mask = np.all(np.equal(np.mod(X, 1), 0), axis=1)
    # Check sum condition
    sum_mask = np.sum(X, axis=1) == n
    return X[integer_mask & sum_mask]
# Example
n = 5
X = np.array([
    [2, 1, 2],
    [1, 1, 1],
    [3, 1, 1],
    [2.5, 1, 1],
    [0, 5, 0]
])
print("Input:\n", X)
print("Valid multinomial rows (sum = 5):\n", multinomial_rows(X, n))
````

````python
#### 100. Compute bootstrapped 95% confidence intervals for the mean of a 1D array X (i.e., resample the elements of an array with replacement N times, compute the mean of each sample, and then compute percentiles over the means). (★★★)
def bootstrap_mean_ci(X, N=10000, ci=95):
    """
    Compute bootstrap confidence interval for mean.
    
    X  : 1D array
    N  : number of bootstrap samples
    ci : confidence level (default 95)
    """
    X = np.asarray(X)
    n = len(X)
    # Generate bootstrap samples (with replacement)
    samples = np.random.choice(X, size=(N, n), replace=True)
    # Compute means
    means = samples.mean(axis=1)
    # Compute confidence interval
    lower = np.percentile(means, (100-ci)/2)
    upper = np.percentile(means, 100-(100-ci)/2)
    return lower, upper
# Example
X = np.random.normal(loc=10, scale=2, size=100)
lower, upper = bootstrap_mean_ci(X)
print("Mean:", np.mean(X))
print("95% CI:", (lower, upper))
````
